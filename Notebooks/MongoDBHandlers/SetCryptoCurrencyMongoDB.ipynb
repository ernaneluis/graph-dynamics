```python
import pymongo
from pymongo import MongoClient

client = MongoClient("129.26.78.131:27017")
```

```python
cryptoCurrenciesDir = "../../data/crypto_currencies/CryptoCurrenciesCSV/"
crypto_currencies = os.listdir(cryptoCurrenciesDir)
all_labels = [a.split("_")[-1][:-4] for a in crypto_currencies]
crypto_currency_label = "BTC"
[a for a in crypto_currencies if crypto_currency_label in a]
```

```python
cryptocurrency_df = pd.read_csv(cryptoCurrenciesDir+"cryptocurrencies_{0}.csv".format(crypto_currency_label),index_col=0,parse_dates=True) 
```

```python
crypto_currencies_data = []
for cryptocurrency_label in all_labels:
    try:
        cryptocurrency_df = pd.read_csv(cryptoCurrenciesDir+"cryptocurrencies_{0}.csv".format(cryptocurrency_label),index_col=0,parse_dates=True) 
        max_date = max(cryptocurrency_df.index)
        min_date = min(cryptocurrency_df.index)
        numberOfDays = len(cryptocurrency_df.index)
        crypto_currencies_data.append({"label":cryptocurrency_label,
                                       "max":max_date.to_datetime(),
                                       "min":min_date.to_datetime(),
                                       "numberOfDays":numberOfDays})
    except:
        print sys.exc_info()
```

```python
db = client["bitcoin"]
db.collection_names()
#db["cryptocurrencies"].drop()
db["cryptocurrencies"].insert(crypto_currencies_data)
```
