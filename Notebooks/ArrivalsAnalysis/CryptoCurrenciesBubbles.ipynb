```python
import os
import numpy as np
import cPickle
import matplotlib.pyplot as plt
import matplotlib
from scipy.optimize import minimize
import pandas as pd

import gpv.finance.bubble_tracker as bt
import datetime

import math

%matplotlib inline
```

```python
CRYPTOCURRENCIES_MARKET_CAP = pd.read_csv("../../data/cryptocurrencies_by_market_cap.csv",parse_dates=True,index_col=0)
TOP_TEN = CRYPTOCURRENCIES_MARKET_CAP["Ticker"].values[:10]

IMPORTANT_DATES = pd.read_csv("../../data/important_dates.csv",parse_dates=True)
CHINA_BAN = pd.to_datetime(IMPORTANT_DATES["China"])[0]
```

```python
ticker = TOP_TEN[0]
crypto_df = pd.read_csv("/home/cesar/Desktop/Doctorado/Projects/BitCoin/Cryptocurrencies/cryptocurrencies_{0}.csv".format(ticker),parse_dates=True,index_col=0)
models = {}
```

```python
crypto_to_fit = crypto_df["high"][datetime.datetime(2017,1,1):]
crypto_to_fit.plot()
```

```python
date0 = datetime.datetime(2017,1,1)
CHINA_BAN = pd.to_datetime(IMPORTANT_DATES["China"])[0]
datef = CHINA_BAN -datetime.timedelta(days=25)

models = {}
paramsLPPL = bt.sornetteWindowLPPL(crypto_to_fit,date0,datef)
crash_date = date0 + datetime.timedelta(days=int(paramsLPPL[3]))
models.update({0: {'window_start': date0, 
                      'window_end': datef,
                      'params': paramsLPPL, 
                      'crash_date': crash_date}})
```

```python
bt.plot_bubbles(models,crypto_to_fit)

plt.plot( np.exp(bt.lppl(paramsLPPL, np.arange(len(crypto_to_fit[date0:])+1000))))
```
